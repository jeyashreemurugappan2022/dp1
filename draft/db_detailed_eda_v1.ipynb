# grouping the data in intervals

```python
s=df.groupby('Age')['Outcome'].sum()
s1=df['Age'].value_counts().sort_index()
d={f'{i+21}-{i+31}':(s[i:i+10].sum()/s1[i:i+10].sum()) for i in range(0,61,10)}
```

```python
s=pd.Series(d)
ax=sns.barplot(x=s.index,y=s.values*100)
for i in ax.containers:
    ax.bar_label(i,)
plt.title('Age Interval vs percentage of diabetes')
```

```python
s=df.groupby('Glucose')['Outcome'].sum()
s1=df['Glucose'].value_counts().sort_index()
d={f'{s1[i:i+17].index[0]}-{s1[i:i+17].index[-1]}':(s[i:i+17].sum()/s1[i:i+17].sum()) for i in range(0,120,17)}
```

```python
s=pd.Series(d)
ax=sns.barplot(x=s.index,y=s.values*100)
for i in ax.containers:
    ax.bar_label(i,)
plt.title('Glucose Interval vs percentage of diabetes')
```
